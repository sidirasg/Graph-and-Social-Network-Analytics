```python
import  networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
g=nx.Graph() #empty graph
# add node 1
g.add_node(1)
g.add_nodes_from([2,3])
g.add_node("spam")        # adds node "spam"
#g.add_nodes_from("spam")  # adds 4
```

```python
g.add_edge(3,'m')

#can also be grown by adding one edge at a time

g.add_edge(1, 2)
e = (2, 3)
g.add_edge(*e)  # unpack edge tuple*
g.add_edges_from([(1, 2), (1, 3)])

#list All edjes and Nodes
list(g.edges)

g.add_edge(3, 'm')

#can also be grown by adding one edge at a time
list(g.node)

g.add_edge(1, 3)
g[1][3]['color'] = "blue"
g.edges[1, 2]['color'] = "red"
#You can get/set the attributes of an edge using subscript notation if the edge already exists.
#Draw the network only the nodes
```

```python
nx.draw_networkx(g,node_color='red',node_size=700)
```
